```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
# file_path = r'5_merge/complete_data.csv'  # Replace with your actual file path
data = pd.read_csv('C:/Users/HP/Desktop/CEDT_Year2/DSDE/Project/DSDE_Project/5_merge/complete_data.csv')
locate = pd.read_excel("C:/Users/HP/Desktop/CEDT_Year2/DSDE/Project/DSDE_Project/5_merge/faculty_location.xlsx")

# Merge the datasets on the 'Faculty' column
df = pd.merge(data, locate, on="Faculty")

# Aggregate data for the heatmap
heatmap_data = df.groupby(['Faculty', 'Year'])['Count'].sum().unstack(fill_value=0)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.title("Heatmap of Thesis Counts by Faculty and Location")
plt.xlabel("Location")
plt.ylabel("Faculty")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
