```python
import pandas as pd
fa = pd.read_csv("faculty.csv")
condition = fa['Bibrecord Organization'] == 'Chulalongkorn University'
fa_cu = fa.where(condition)
fa_cu = fa_cu.dropna()
fa_cu.to_csv("faculty_cu.csv",index=False)
```
